```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
#How do we get the data
```

```python
digits
```

```python
X = digits['data']
y = digits['target']
```

```python
X[0]
```

```python
y[0:20]
```

```python


plt.gray()
select = 5 #corresopnds to data in X[number]
plt.matshow(digits.images[select])
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
scaledX = ss.fit_transform(X)
```

```python
#Do the PCA
pca = PCA()
```

```python
pca_digits = pca.fit_transform(scaledX)
```

```python
pca.explained_variance_ratio_[0:10]
```

```python
pca_df = pd.DataFrame(pca_digits, columns = ['PC'+str(i) for i in range(1,len(pca_digits[0])+1)])
pca_df
```

```python
plt.figure(figsize = [8,8])
colors = ['cornflowerblue', 'slategrey','lightsalmon',
          'violet','forestgreen','firebrick', 'cyan',
         'lawngreen','thistle','black']
sns.scatterplot(data = pca_df, x = 'PC1',y = 'PC2', hue = y, palette = colors, alpha = 0.8)
plt.legend(bbox_to_anchor = [1,0.75], title = 'Actual digit', frameon = False, fontsize = 14,
          title_fontsize = 16)
plt.xlabel('PC1', fontsize = 14)
plt.ylabel('PC2', fontsize = 14)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
#do the tsne, use random state 99
tsne = TSNE(random_state=99) # default perplexity = 10
```

```python
# in general, we also scale data for TSNE
tsne_digits = tsne.fit_transform(X)
```

```python
tsne_df = pd.DataFrame(tsne_digits, columns = ['TSNE1','TSNE2'])
```

```python
plt.figure(figsize = [8,8])
colors = ['cornflowerblue', 'slategrey','lightsalmon',
          'violet','forestgreen','firebrick', 'cyan',
         'lawngreen','thistle','black']
sns.scatterplot(data = tsne_df, x = 'TSNE1',y = 'TSNE2', hue = y, palette = colors, alpha = 0.8)
plt.legend(bbox_to_anchor = [1,0.75], title = 'Actual digit', frameon = False, fontsize = 14,
          title_fontsize = 16)
plt.xlabel('TSNE1', fontsize = 14)
plt.ylabel('TSNE2', fontsize = 14)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
nines = tsne_df[tsne_df['target'] == 9]
nines_int = nines[nines['TSNE2'] > 20]
```

```python
for i in nines_int.index:
    plt.gray()
    plt.matshow(digits.images[i])
    plt.show()
```
